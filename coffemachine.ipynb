```python
import os


with open("student.txt", "w") as file:
    file.write("Dhruv\n")
    file.write("Marks: 90\n")
    file.write("Course: Python")

print("Data written successfully")



with open("student.txt", "r") as file:
    content = file.read()
    print("\nComplete File Content:")
    print(content)



with open("student.txt", "r") as file:
    print("\nReading line by line:")
    
    for line in file:
        print(line.strip())



with open("student.txt", "r") as file:
    print("\nFirst Line:")
    print(file.readline())



with open("student.txt", "a") as file:
    file.write("\nGrade: A")

print("\nNew data appended")


# READ AFTER APPEND
with open("student.txt", "r") as file:
    print("\nUpdated File:")
    print(file.read())


# CHECK FILE EXISTS
if os.path.exists("student.txt"):
    print("\nFile exists")


# FILE POINTER POSITION
with open("student.txt", "r") as file:
    
    print("\nCursor Position:")
    print(file.tell())

    file.read(5)

    print("\nCursor Position After Reading 5 Characters:")
    print(file.tell())

    file.seek(0)

    print("\nCursor Reset:")
    print(file.tell())


# CREATE NEW FILE
try:
    with open("newfile.txt", "x") as file:
        print("\nNew file created")

except FileExistsError:
    print("\nFile already exists")


# DELETE FILE
choice = input("\nDo you want to delete newfile.txt? (yes/no): ")

if choice == "yes":
    
    if os.path.exists("newfile.txt"):
        os.remove("newfile.txt")
        print("File deleted")

    else:
        print("File not found")
```

```python
import requests


API_KEY = "229bb6342139de7200c4592ed204b1d8"       
CITY    = "London"                  
UNITS   = "metric"                  
BASE_URL = "https://api.openweathermap.org/data/2.5/weather"


params = {
    "q":     CITY,
    "appid": API_KEY,
    "units": UNITS
}



response = requests.get(BASE_URL, params=params)

# Check if request was successful
if response.status_code != 200:
    print(f"❌ Error {response.status_code}: {response.json().get('message')}")
    exit() # Exit the script if the API call fails

data = response.json()   # Convert JSON response → Python dict

\
rows = [
    ("City",             data["name"]),
    ("Country",          data["sys"]["country"]),
    ("Weather",          data["weather"][0]["description"].title()),
    ("Temperature (°C)", f"{data['main']['temp']} °C"),
    ("Feels Like",       f"{data['main']['feels_like']} °C"),
    ("Min Temp",         f"{data['main']['temp_min']} °C"),
    ("Max Temp",         f"{data['main']['temp_max']} °C"),
    ("Humidity",         f"{data['main']['humidity']} %"),
    ("Pressure",         f"{data['main']['pressure']} hPa"),
    ("Wind Speed",       f"{data['wind']['speed']} m/s"),
    ("Visibility",       f"{data.get('visibility', 'N/A')} m"),
    ("Cloudiness",       f"{data['clouds']['all']} %"),
]

# ─────────────────────────────────────────
# STEP 4 — DISPLAY IN ROWS & COLUMNS
# ─────────────────────────────────────────
print("\n" + "=" * 40)
print(f"  🌤  WEATHER REPORT — {CITY.upper()}")
print("=" * 40)
print(f"{'FIELD':<20} {'VALUE':<20}")
print("-" * 40)

for field, value in rows:
    print(f"{field:<20} {value:<20}")

print("=" * 40)
```
