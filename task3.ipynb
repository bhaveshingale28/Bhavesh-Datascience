```python
matches_per_season = pd.DataFrame(
    df.groupby('season').size().reset_index(name='Total_Matches')
)
print(matches_per_season)
```

```python
most_wins = pd.DataFrame(
    df['winner'].value_counts().reset_index()
)

most_wins.columns = ['Team', 'Wins']

print("\nTeam With Most Wins Overall")
print(most_wins.head(1))
```

```python
venue_name = "M Chinnaswamy Stadium"

venue_matches = pd.DataFrame(
    df[df['venue'] == venue_name]
)

print(f"Matches Played at {venue_name}")
print(venue_matches[['season', 'team1', 'team2', 'winner']])

top_players = pd.DataFrame(
    df['player_of_match'].value_counts().head(5).reset_index()
)

top_players.columns = ['Player', 'Awards']

print("\nTop 5 Players With Most Player of the Match Awards")
print(top_players)

print("\nNull Values in Dataset")
print(df.isnull().sum())

clean_df = df.dropna()

print("\nDataset Shape After Removing Null Values")
print(clean_df.shape)
```
